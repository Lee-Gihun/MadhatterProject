```python
import json

with open('datasets/csv/id_to_key' + '.json', encoding='utf-8') as g:
    id_to_key_data = json.load(g)

to_dict = {}
to_dict = id_to_key_data
```

```python
def dict_to_df(d, col1, col2):
    result = pd.Series(d, name=col2)
    result.index.name = col1
    result = result.reset_index()
    return result
```

```python
s = pd.Series(to_dict, name='Key')
s.index.name = 'Champion'
df = s.reset_index()
df
```

```python
def map_champ_id_to_key(champion_edges, id_key_tb):
    champion_edges['Source'] = champion_edges['Source'].map(id_key_tb.set_index('Champion')['Key']) 
    champion_edges['Target'] = champion_edges['Target'].map(id_key_tb.set_index('Champion')['Key']) 
    return champion_edges
```

```python
def get_indices_value(a):
    d = {}
    for i in range(len(a)):
        if a[i] > 0:
            d[i] = a[i]
    return d
```

```python
def filter_user_tfidf(user_vector):
    to_original_filtered = get_indices_value(user_vector)
    return to_original_filtered
```

```python
def get_recommendation(filtered_user_df, weight_edges):
    f = dict_to_df(filtered_user_df, 'Id', 'tfidf')
    filtered_edges = weight_edges[weight_edges['Source'].isin(f['Id'])].copy()
    filtered_edges = filtered_edges[~filtered_edges['Target'].isin(f['Id'])].copy()
    filtered_edges['Weight*Tfidf'] = filtered_edges.apply(lambda row: filtered_user_df.loc[filtered_user_df['Id'] == row['Source'].item(), 'tfidf'].item() * row['Weight'], axis=1)
    filtered_edges['sum_weighted_edges'] = filtered_edges.groupby(['Target'])['Weight*Tfidf'].transform(lambda x: x.sum())
    filtered_edges.sort_values(by='sum_weighted_edges', ascending=False, inplace=True)
    recommendation_lists_ids = filtered_edges[['Target','sum_weighted_edges']].copy()
    recommendation_lists_ids.drop_duplicates(subset='Target', inplace=True)
    return recommendation_lists_ids
```

```python
def map_key_to_champ_id(recomm_lists_id, id_key_tb):
    result = recomm_lists_id.copy()
    result['Target'] = result['Target'].map(id_key_tb.set_index('Key')['Champion']) 
    return result
```
